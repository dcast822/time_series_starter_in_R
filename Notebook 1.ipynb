### Boxplot by Cycle

```
boxplot(AirPassengers~cycle(AirPassengers, ylab = "Passenger Numbers(1000s)",
 main = 'Monthly air passengers boxplot from 1949-1960'))
```

Plots indicate the highest ticket sales are in the summer months of June, July, August

### Decompose Data into 4 Components

```
tsdata <- ts(AirPassengers, frequency = 12)
ddata <- decompose(tsdata, "multiplicative")
```

```
plot(ddata)
```

### Trendline for original dataset

```
plot(AirPassengers)

abline(reg=lm(AirPassengers~time(AirPassengers)))
```

## Arima Modeling

```
arimamodel <-auto.arima(AirPassengers)
arimamodel
```

### Plotting of Residuals

```
plot.ts(arimamodel$residuals)
```

### Forecast next ten years

```
arimaforecast <- forecast(arimamodel, level = c(95), h = 10*12)
plot(arimaforecast)
```

## Validate Model using lag values

```
Box.test(arimamodel$resid, lag = 5, type='Ljung-Box')
```

```
Box.test(arimamodel$resid, lag = 10, type = "Ljung-Box")
```

```
Box.test(arimamodel$resid, lag = 15, type = 'Ljung-Box')
```
